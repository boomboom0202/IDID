# Split

```python
metadata = []
for filepath in data_dirpath.glob('**/images/**/*.*'):
    split = filepath.parent.name
    id_ = filepath.stem
    if id_[-1] in {'v', 'd', 'h'}:
        id_ = id_[:-1]
        
    label_counts = {0: 0, 1: 0, 2: 0, 3: 0, 4: 0}
    label_filepath = (filepath.parents[2] / 'labels' / split / filepath.stem).with_suffix('.txt')
    if label_filepath.exists():
        labels = pd.read_csv(label_filepath, sep=' ', header=None, names=['class', 'x', 'y', 'w', 'h'])
        label_counts = label_counts | labels['class'].value_counts().to_dict()
    
    metadata.append(
        {
            'filepath': filepath,
            'name': filepath.stem,
            'id': id_,
            'split': split,
        } | label_counts
    )
df_metadata = pd.DataFrame(metadata)
df_metadata
```

```python
df_metadata = df_metadata[df_metadata['split'].isin({'train', 'val'})]
df_metadata
```

```python
for i in range(5):
    print(df_metadata[i].value_counts().to_dict())
```

```python
df_metadata_unique = df_metadata[['id', 1, 2, 4]].drop_duplicates(subset=['id'])
df_metadata_unique[[1, 2, 4]] = df_metadata_unique[[1, 2, 4]] > 0
df_metadata_unique
```

```python
kfold = MultilabelStratifiedKFold(n_splits=5, shuffle=True, random_state=42)
df_metadata_unique['fold'] = -1
for fold, (train_val_idx, test_idx) in enumerate(kfold.split(df_metadata_unique, df_metadata_unique[[1, 2, 4]], df_metadata_unique['id'])):
    df_metadata_unique.iloc[test_idx, -1] = fold
df_metadata_unique
```

```python
for fold in range(5):
    print(df_metadata_unique[df_metadata_unique['fold'] == fold][[1, 2, 4]].sum().to_dict())
```
